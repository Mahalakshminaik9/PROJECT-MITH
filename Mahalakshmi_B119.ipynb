Data PreProcessing

```python
from sklearn.model_selection import train_test_split
```

```python
X = traindata_tax.drop(['PropertyEvaluationvalue'], axis = 1)
y = traindata_tax['PropertyEvaluationvalue']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

Training Model

```python

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Lasso
from sklearn import metrics

```

Standard Scaling

```python
num_attr = X_train.select_dtypes(['int64']).columns
num_attr
```

```python
scaler = StandardScaler()
```

```python
X_train[num_attr]=scaler.fit_transform(X_train[num_attr])
```

```python
X_test[num_attr]=scaler.transform(X_test[num_attr])
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
lin_reg_model = LinearRegression()
```

```python
lin_reg_model.fit(X_train,y_train)
```

Model Evaluation

```python
training_data_prediction = lin_reg_model.predict(X_train)
```

```python
error_score = metrics.r2_score(y_train, training_data_prediction)
print("R squared : ", error_score)
```

```python
plt.scatter(y_train, training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title(" Actual Prices vs Predicted Prices")
plt.show()
```

```python
lass_reg_model = Lasso()
```

```python
lass_reg_model.fit(X_train,y_train)
```

```python
training_data_prediction = lass_reg_model.predict(X_train)
```

```python
error_score = metrics.r2_score(y_train, training_data_prediction)
print("R squared Error : ", error_score)
```

```python
plt.scatter(y_train, training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title(" Actual Prices vs Predicted Prices")
plt.show()
```

Test Data

```python
y_test_pred = lin_reg_model.predict(X_test)
```

```python
error_score = metrics.r2_score(y_test, y_test_pred )
print("R squared Error : ", error_score)
```
